```python
import pandas as pd
from prettytable import PrettyTable
import matplotlib.pyplot as plt
import seaborn as sns
apart = pd.read_csv('/content/sample_data/apartments_for_rent_classified_10K.csv', encoding='latin1', sep=';')
apart.describe()
```

**1.** **Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka**

```python
# Memilih hanya kolom numerik
df_dataApart = apart.select_dtypes(include=['number'])

# Menghitung statistik deskriptif
statistik = df_dataApart.describe()

# Mengambil statistik yang diperlukan
rata_rata = statistik.loc['mean']
standarDeviasi = statistik.loc['std']
nilaiMinimum = statistik.loc['min']
q1 = df_dataApart.quantile(0.25)
median = statistik.loc['50%']
q3 = df_dataApart.quantile(0.75)
nilaiMaksimum = statistik.loc['max']

# Membuat DataFrame untuk hasil
tableStatistikApart = pd.DataFrame({
    'Rata-rata': rata_rata,
    'Standar Deviasi': standarDeviasi,
    'Nilai Minimum': nilaiMinimum,
    'Q1': q1,
    'Median (Q2)': median,
    'Q3': q3,
    'Nilai Maksimum': nilaiMaksimum})

# Menampilkan tabel
print(tableStatistikApart)
```

**3.** **Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.**

```python
dataApart = {}

dataApart['Jumlah Record'] = apart.shape[0]
dataApart['Jumlah Attribute'] = apart.shape[1]
dataApart['Nama Attribute'] = apart.columns.tolist()
dataApart['Jumlah Attribute Dengan Tipe Angka'] = apart.select_dtypes(include=['int64', 'float64']).shape[1]

dictApart = list(dataApart.keys())

table = PrettyTable()
table.add_column(dictApart[0], [dataApart['Jumlah Record']])
table.add_column(dictApart[1], [dataApart['Jumlah Attribute']])
table.add_column(dictApart[2], ['\n'.join(dataApart['Nama Attribute'])])
table.add_column(dictApart[3], [dataApart['Jumlah Attribute Dengan Tipe Angka']])
print(table)
```

**4.** **Tampilkan jumlah label dari semua attribute yang bernilai object**

```python
# Memilih hanya kolom bertipe objek
apart_objek = apart.select_dtypes(include=['object'])

# Menghitung jumlah label unik untuk setiap kolom
jumlah_label = apart_objek.nunique()

# Membuat tabel menggunakan PrettyTable
table = PrettyTable()

# Menambahkan kolom ke tabel
table.field_names = ["Nama Atribut", "Jumlah Label"]

# Menambahkan data jumlah label unik ke dalam tabel
for column, count in jumlah_label.items():
    table.add_row([column, count])

print(table)
```

**5.** **Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.**

```python
# Menghitung jumlah setiap kategori bedrooms
dataBedroom = apart['bedrooms'].value_counts().reset_index()
dataBedroom.columns = ['bedrooms', 'count']

# Membuat grafik batang
plt.figure(figsize=(10, 6))
sns.barplot(data=dataBedroom, x='bedrooms', y='count')
plt.title('Jumlah Penyewaan Berdasarkan Jumlah Kamar Tidur')
plt.xlabel('Jumlah Kamar Tidur')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

**6.** **Tampilkan korelasi dataset menggunakan heatmap**

```python
# Memilih hanya kolom numerik
kolomNumerik = apart.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antar atribut
korelasi = kolomNumerik.corr()

# Membuat heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(korelasi, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar=True)
plt.title('Heatmap Korelasi Atribut')
plt.show()
```
